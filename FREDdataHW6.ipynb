```python
url_base_one = "https://api.stlouisfed.org/fred/series/observations?series_id={}&api_key=22ee7a76e736e32f54f5df0a7171538d&file_type=json"

file_base_one = "FRED_data_{}.json"

fredSeries = ['GFDEGDQ188S','IRLTLT01DEM156N','MSPUS','QGBN628BIS']

for i in fredSeries:
  print(i)
  URL = url_base_one.format(i)
  print(URL)
  data = requests.get(URL).json()
  print(data)

  fileName_one = file_base_one.format(i)
  print(fileName_one)

  with open(fileName_one, 'w', encoding='utf-8') as f:
     json.dump(data, f, ensure_ascii=False, indent=4)

  files.download(fileName_one)
```
